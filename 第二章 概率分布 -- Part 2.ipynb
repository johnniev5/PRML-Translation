#### 2.3.1 条件高斯分布

多变量高斯分布的一个重要特性，就是如果两变量的集合是联合概率，那么一个集合的条件分布在另一个上的条件也是高斯。类似地，相应的边缘分布也是高斯。

考虑第一个条件分布的例子。假设$\mathbf{x}$是一个$D$维向量的高斯分布$\mathcal{N}(\mathbf{x} \mid \mathbf{\mu, \Sigma})$，现在我们将$\mathbf{x}$分成两个联合子集$\mathbf{x}_a$和$\mathbf{x}_b$。不失一般性，我们可以将$\mathbf{x}_a$看作是$\mathbf{x}$的$M$个部分，相应的$\mathbf{x}_b$是剩余的$D-M$个部分，所以有

$$\mathbf{x} = \begin{pmatrix}
\mathbf{x}_a \\
\mathbf{x}_b
\end{pmatrix}$$

同样，我么定义相应部分的均值向量$\mathbf{\mu}$如下

$$\mathbf{x} = \begin{pmatrix}
\mathbf{\mu}_a \\
\mathbf{\mu}_b
\end{pmatrix}$$

其协方差矩阵$\mathbf{\Sigma}$如下

$$\mathbf{\Sigma} = \begin{pmatrix}
\mathbf{\Sigma}_{aa}  & \mathbf{\Sigma}_{ab}\\
\mathbf{\Sigma}_{ba} & \mathbf{\Sigma}_{bb}
\end{pmatrix}$$

需要注意的是协方差矩阵的对称性$\mathbf{\Sigma}^T = \mathbf{\Sigma}$应用到$\mathbf{\Sigma}_{aa}$和$\mathbf{\Sigma}_{bb}$是对称的，而$\mathbf{\Sigma}_{ba} = \mathbf{\Sigma}_{ab}^T$。

在许多条件里，使用协方差矩阵的逆将会更方便

$$\Lambda \equiv \mathbf{\Sigma}^{-1}$$

这被称为精度矩阵。事实上，我们会看到高斯分布的某些性质很自然地就用协方差来表示，而另一些性质则用精度来表示更简单。我们因此会引入精度矩阵的分解形式

$$\mathbf{\Lambda} = \begin{pmatrix}
\mathbf{\Lambda}_{aa}  & \mathbf{\Lambda}_{ab}\\
\mathbf{\Lambda}_{ba} & \mathbf{\Lambda}_{bb}
\end{pmatrix}$$

与2.65的向量$\mathbf{x}$相对应。因为一个对称矩阵的逆还是对称的，我们看到$\mathbf{\Lambda}_{aa}$与$\mathbf{\Lambda}_{bb}$是对称的，而$\mathbf{\Lambda}_{ba}^T = \mathbf{\Lambda}_{ab}$。需要强调的是，例如，$\mathbf{\Lambda}_{aa}$并不是简单地由$\mathbf{\Lambda}_{aa}$的逆得到的。事实上，我们应该简短地验证一下分块矩阵的逆与逆矩阵的分块之间的关系。

让我们先找一个条件分布$p(\mathbf{\mu}_a \mid \mathbf{\mu}_b)$的表达式。从概率的乘积规则，我们看到，只需将$\mathbf{\mu}_b$固定到观测值，并对所得表达式进行规范化，就可以从联合分布$p(\mathbf{x}) = p(\mathbf{x}_a, \mathbf{x}_b)$评估这种条件分布，从而获得在$\mathbf{x}_a$上的有效概率分布。

我们不直接用标准项，我们可以从2.44高斯分布的指数项中的二次项来得到更有效的解，然后再计算结束时，再加上标准化系数。我们可以使用2.65，2.66和2.69中的分块，可以得到如下

$$\begin{align*}
& (\mathbf{x} - \mathbf{\mu})^T\mathbf{\Sigma}^{-1}(\mathbf{x} - \mathbf{\mu}) = \\
& -\frac{1}{2}(\mathbf{x}_a - \mathbf{\mu}_a)^T\mathbf{\Lambda}_{aa}(\mathbf{x}_a - \mathbf{\mu}_a) - \frac{1}{2}(\mathbf{x}_a - \mathbf{\mu}_a)^T\mathbf{\Lambda}_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) \\
& -\frac{1}{2}(\mathbf{x}_b - \mathbf{\mu}_b)^T\mathbf{\Lambda}_{ba}(\mathbf{x}_a - \mathbf{\mu}_a) - \frac{1}{2}(\mathbf{x}_a - \mathbf{\mu}_a)^T\mathbf{\Lambda}_{bb}(\mathbf{x}_b - \mathbf{\mu}_b)
\end{align*}$$

我们看到$\mathbf{x}_a$的函数，仍然是一个二次型，因此相应的条件分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$是一个高斯。因为这个分布完全以其均值和协方差为特征，我们的目标是通过检查2.70，确定$p(\mathbf{\mu}_a \mid \mathbf{\mu}_b)$的均值和协方差的表达式。

这是一个相当普通的与高斯分布相关的操作，有时称作"完全平方"，它定义了高斯分布中的二次型指数项，我们也需要定义相应的均值和协方差。这样的问题可以直接通过通用高斯分布$\mathcal{N(\mathbf{x} \mid \mathbf{\mu}, \mathbf{\Sigma})}$的指数项直接解决，可以写成如下

$$-\frac{1}{2}(\mathbf{x} - \mathbf{\mu})^T\mathbf{\Sigma}^{-1}(\mathbf{x} - \mathbf{\mu}) = -\frac{1}{2}\mathbf{x}^T\mathbf{\Sigma}^{-1}\mathbf{x} + \mathbf{x}^T\mathbf{\Sigma}^{-1}\mathbf{\mu} + \text{const}$$

其中，$\text{const}$表示的与$\mathbf{x}$独立的项，这里我们也使用了$\mathbf{\Sigma}$的对称性。因此如果我们取二次型，并用2.71右边给出的形式来表示，然后我们可以直接使得$\mathbf{x}$找那个二阶项的系数矩阵对应的逆矩阵$\mathbf{\Sigma}^{-1}$等于$\mathbf{x}$的线性项中的系数$\mathbf{\Sigma}^{-1}\mu$，由此我们可以得到$\mathbf{\mu}$。

现在让我们将这个流程应用到条件高斯分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$，正如2.70中给出的指数中的二次型。我们会使用相应地$\mathbf{\mu}_{a \mid b}$和$\mathbf{\Sigma}_{a \mid b}$表示均值和方差。考虑2.70中与$\mathbf{x}_a$相关的函数，而这里的$\mathbf{x}_b$被看视作是常量。如果我们取所有$\mathbf{x}_a$的二次项，我们有

$$-\frac{1}{2}\mathbf{x}_a^T\mathbf{\Lambda}_{aa}\mathbf{x}_a$$

从中我们也可以很快地总结出协方差（逆精度）$p(\mathbf{x}_a \mid \mathbf{x}_b)$给出如下

$$\mathbf{\Sigma}_{a \mid b} = \mathbf{\Lambda}_{aa}^{-1}$$

我们现在考虑2.70中的所有线性项

$$\mathbf{x}_a\{\mathbf{\Lambda}_{aa}\mathbf{\mu}_a - \mathbf{\Lambda}_{ab}(\mathbf{x}_b - \mathbf{\mu}_b)\}$$

其中我们已经有$\mathbf{\Lambda}_{ba}^T = \mathbf{\Lambda}_{ab}$。从我们讨论的2.71中的通用形式，表达式中$\mathbf{x}_a$的系数必须等于$\mathbf{\Sigma}_{a \mid b}\mathbf{\mu}_{a \mid b}$，因此

$$\begin{align*}\mathbf{\mu}_{a \mid b}
&= \mathbf{\Sigma}_{a \mid b}\{\mathbf{\Lambda}_{aa}\mathbf{\mu}_a - \mathbf{\Lambda}_{ab}(\mathbf{x}_b - \mathbf{\mu}_b)\} \\
&= \mathbf{\mu}_a - \mathbf{\Lambda}_{aa}^{-1}\mathbf{\Lambda}_{ab}(\mathbf{x}_b - \mathbf{\mu}_b)
\end{align*}$$

其中这里我们使用了2.73。

2.73和2.75的结果被表达成原始联合分布$p(\mathbf{x}_a, \mathbf{x}_b)$中分块精度矩阵。我们也以相应的分块协方差矩阵来表达这些结果。如果要这样的话，我们使用下面的恒等式求分块矩阵的逆矩阵

$$\begin{pmatrix}
\mathbf{A} & \mathbf{B} \\
\mathbf{C} & \mathbf{D}
\end{pmatrix}^{-1} =
\begin{pmatrix}
\mathbf{M} & -\mathbf{MBD}^{-1} \\
-\mathbf{D}^{-1}\mathbf{CM} & \mathbf{D}^{-1} + \mathbf{D}^{-1}\mathbf{CMBD}^{-1}
\end{pmatrix}$$

其中我们定义

$$\mathbf{M} = (\mathbf{A} - \mathbf{B}\mathbf{D}^{-1}\mathbf{C})^{-1}$$

量$\mathbf{M}^{-1}$被称为2.76左侧矩阵相应的子矩阵$\mathbf{D}$的舒尔布。使用如下定义

$$\begin{pmatrix}
\mathbf{\Sigma}_{aa} & \mathbf{\Sigma}_{ab} \\
\mathbf{\Sigma}_{ba} & \mathbf{\Sigma}_{bb}
\end{pmatrix}^{-1} =
\begin{pmatrix}
\mathbf{\Lambda}_{aa} & \mathbf{\Lambda}_{ab} \\
\mathbf{\Lambda}_{ba} & \mathbf{\Lambda}_{bb}
\end{pmatrix}$$

使用2.76，我们有

$$\mathbf{\Lambda}_{aa} = (\mathbf{\Sigma}_{aa} - \mathbf{\Sigma}_{ab}\mathbf{\Sigma}_{bb}^{-1}\mathbf{\Sigma}_{ba})^{-1}$$

$$\mathbf{\Lambda}_{ab} = -(\mathbf{\Sigma}_{aa} - \mathbf{\Sigma}_{ab}\mathbf{\Sigma}_{bb}^{-1}\mathbf{\Sigma}_{ba})^{-1}$$

从中我们得到下面条件分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$的均值和方差

$$\mathbf{\mu}_{a \mid b} = \mathbf{\mu}_a + \Sigma_{ab}\Sigma_{bb}^{-1}(\mathbf{x}_b - \mathbf{\mu}_b)$$

$$\mathbf{\Sigma}_{a \mid b} = \mathbf{\Sigma}_{aa} - \mathbf{\Sigma}_{ab}\mathbf{\Sigma}_{bb}^{-1}\mathbf{\Sigma}_{ba}$$

比较2.73和2.82，我们看到条件分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$使用分块精度矩阵来时表示要比表示成分块协方差矩阵来的简单。需要注意的是条件分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$的均值，如2.81给出的，是一个$\mathbf{x}_b$的线性函数，而2.82给出的协方差是与$\mathbf{x}_a$无关的。这表示的是一个线性高斯模型的例子。

#### 2.3.2 边缘高斯分布

我们已经看到如果一个联合分布$p(\mathbf{x}_a, \mathbf{x}_b)$是高斯，那么条件分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$也是高斯。现在我们转到边缘高斯的讨论

$$p(\mathbf{x}_a) = \int p(\mathbf{x}_a \mid \mathbf{x}_b)d\mathbf{x}_b$$

同样，我们可以看到它也是一个高斯。再次，我们的策略还是将有效评估聚焦于联合分布的指数项中的二次型，然后，得到边远分部的均值和协方差。

联合分布的二次型可以表达成2.70中的分块精度形式。因为我们的目标是整合$\mathbf{x}_b$，这是很容易实现的，首先考虑设计$\mathbf{x}_b$的项，然后完全平方项以便于集成。只将包含$\mathbf{x}_b$的这些项取出，我们有

$$-\frac{1}{2}\mathbf{x}_b\mathbf{\Lambda}_{bb}\mathbf{x}_b + \mathbf{x}_b^T\mathbf{m} = -\frac{1}{2}(\mathbf{x}_b - \mathbf{\Lambda}_{bb}^{-1}\mathbf{m})^T\mathbf{\Lambda}_{bb}(\mathbf{x}_b - \mathbf{\Lambda}_{bb}^{-1}\mathbf{m}) + \frac{1}{2}\mathbf{m}^T\mathbf{\Lambda}_{bb}^{-1}\mathbf{m}$$

其中，我们有如下定义

$$\mathbf{m} = \mathbf{\Lambda}_{bb}\mathbf{\mu}_b - \mathbf{\Lambda}_{ba}(\mathbf{x}_a - \mathbf{\mu}_a)$$

$$\mathbf{\mu}_b = \mathbf{\Lambda}_{bb}^{-1}\mathbf{m}$$

我们看到在2.84右侧的相应的第一项是与$\mathbf{x}_b$相关的高斯分布的标准二次型，加上一项不依赖于$\mathbf{x}_b$的项（但依赖于$\mathbf{x}_a$）。因此，当我们取二次型指数时，我们看到2.83中对于$\mathbf{x}_b$的整合将取如下形式

$$\int \text{exp}\left\{-\frac{1}{2}(\mathbf{x}_b - \mathbf{\Lambda}_{bb}^{-1}m)^T\mathbf{\Lambda}_{bb}(\mathbf{x}_b - \mathbf{\Lambda}_{bb}^{-1}m)\right\}d\mathbf{x}_b$$

这个积分很容易实现，注意到它是非标准化高斯的积分。所以结果是归一化系数的倒数。我们从2.43知道了标准化高斯，它的系数是独立于均值的，值依赖于协方差矩阵的行列式。因此，通过完全平方代表$\mathbf{x}_b$，我们可以整合$\mathbf{x}_b$，并且2.84左边贡献的唯一一个依赖于$\mathbf{x}_a$的项时2.84右边最后一项中的由2.85中$\mathbf{m}$给出。考虑依赖于$\mathbf{x}_a$的剩余项，我们有

$$\begin{align*}
&\frac{1}{2}[\mathbf{\Lambda}_{bb}\mathbf{\mu}_b - \mathbf{\Lambda}_{ba}(\mathbf{x}_a - \mathbf{\mu}_a)]^T\mathbf{\Lambda}_{bb}^{-1}[\mathbf{\Lambda}_{bb}\mathbf{\mu}_b - \mathbf{\Lambda}_{ba}(\mathbf{x}_a - \mathbf{\mu}_a)] - \frac{1}{2}\mathbf{x}_a^T\mathbf{\Lambda}_{aa}\mathbf{x}_a + \mathbf{x}_a^T(\mathbf{\Lambda}_{aa}\mathbf{\mu}_a + \mathbf{\Lambda}_{ab}\mathbf{\mu}_b) + \text{const} \\
&= 
-\frac{1}{2}\mathbf{x}_a^T(\mathbf{\Lambda}_{aa} - \mathbf{\Lambda}_{ab}\mathbf{\Lambda}_{bb}^{-1}\mathbf{\Lambda}_{ba})\mathbf{x}_a + \mathbf{x}_a(\mathbf{\Lambda}_{aa} - \mathbf{\Lambda}_{ab}\mathbf{\Lambda}_{bb}^{-1}\mathbf{\Lambda}_{ba})\mathbf{\mu}_a + \text{const}
\end{align*}$$

这里的$\text{const}$表示的是与$\mathbf{x}_a$无关的量。再次与2.71进行比较，我们看到$p(\mathbf{x}_a)$的边缘分布的协方差如下给出

$$\mathbf{\Sigma}_a = (\mathbf{\Lambda}_{aa} - \mathbf{\Lambda}_{ab}\mathbf{\Lambda}_{bb}^{-1}\mathbf{\Lambda}_{ba})^{-1}$$

类似地，均值由如下给出

$$\mathbf{\Sigma}_a(\mathbf{\Lambda}_{aa} - \mathbf{\Lambda}_{ab}\mathbf{\Lambda}_{bb}^{-1}\mathbf{\Lambda}_{ba})\mathbf{\mu}_a = \mathbf{\mu}_a$$

其中我们用了2.88。2.88中国的协方差矩阵使用了2.69中的精度矩阵来时表示。正如我们在条件分布中做的那样，我们可以重写2.67相应的分块协方差矩阵。这些分块矩阵相关

$$\begin{pmatrix}
\mathbf{\Lambda}_{aa} & \mathbf{\Lambda}_{ab} \\
\mathbf{\Lambda}_{ba} & \mathbf{\Lambda}_{bb}
\end{pmatrix}^{-1} =
\begin{pmatrix}
\mathbf{\Sigma}_{aa} & \mathbf{\Sigma}_{ab} \\
\mathbf{\Sigma}_{ba} & \mathbf{\Sigma}_{bb}
\end{pmatrix}$$

使用2.76有，我们然后有

$$(\mathbf{\Lambda}_{aa} - \mathbf{\Lambda}_{ab}\mathbf{\Lambda}_{bb}^{-1}\mathbf{\Lambda}_{ba})^{-1} = \mathbf{\Sigma}_{aa}$$

因此我们直觉地得到满足边缘分布$p(\mathbf{x}_a)$的结果的均值和协方差

$$\mathbb{E}[\mathbf{x}_a] = \mathbf{\mu}_a$$

$$\text{cov}[\mathbf{x}_a] = \mathbf{\Sigma}_{aa}$$

我们看到边缘分布，均值和协方差都简单地表达成分块协方差矩阵项，相比条件分布，它使用的是分块精度矩阵产生的简单表达式。

分块高斯的边缘和条件分布的结果总结如下：

> **<font color='red'>分块高斯</font>**

> 给出联合高斯分布$\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}, \mathbf{\Sigma})$，有$\mathbf{\Lambda} \equiv \mathbf{\Sigma}^{-1}$，和如下
$$\mathbf{x} = \begin{pmatrix}
\mathbf{x}_a \\
\mathbf{x}_b
\end{pmatrix}, \hspace{0.5cm}
\mathbf{\mu} = \begin{pmatrix}
\mathbf{\mu}_a \\
\mathbf{\mu}_b
\end{pmatrix}$$
$$\mathbf{\Sigma} = \begin{pmatrix}
\mathbf{\Sigma}_{aa}  & \mathbf{\Sigma}_{ab}\\
\mathbf{\Sigma}_{ba} & \mathbf{\Sigma}_{bb}
\end{pmatrix}, \hspace{0.5cm}
\mathbf{\Lambda} = \begin{pmatrix}
\mathbf{\Lambda}_{aa}  & \mathbf{\Lambda}_{ab}\\
\mathbf{\Lambda}_{ba} & \mathbf{\Lambda}_{bb}
\end{pmatrix}$$

> 条件分布：
$$p(\mathbf{x}_a \mid \mathbf{x}_b) = \mathcal{N}(\mathbf{x} \mid \mathbf{\mu}_{a \mid b}, \mathbf{\Lambda}_{aa}^{-1})$$
$$\mathbf{\mu}_{a \mid b} = \mathbf{\mu}_a - \mathbf{\Lambda}_{aa}^{-1}\mathbf{\Lambda}_{ab}(\mathbf{x}_b - \mathbf{\mu}_b)$$

> 边缘分布：
$$p(\mathbf{x}_a) = \mathcal{N}(\mathbf{x}_a \mid \mathbf{\mu}_a, \mathbf{\Sigma}_{aa})$$

#### 2.3.3 高斯变量中的贝叶斯定理

在2.3.1和2.3.2节，我们考虑了高斯$p(\mathbf{x})$，将向量$\mathbf{x}$划分成两个字向量$\mathbf{x} = (\mathbf{x}_a, \mathbf{x}_b)$，然后找到条件分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$和边远分部$p(\mathbf{x}_a)$的表达式。我们注意到条件分布$p(\mathbf{x}_a \mid \mathbf{x}_b)$的均值是$\mathbf{x}_b$的线性函数。这里我们将假设，我们有一个高斯边缘分布$p(\mathbf{x})$，和一个高斯条件分布$p(\mathbf{y} \mid \mathbf{x})$，它的均值是$\mathbf{x}$的线性函数，协方差是独立于$\mathbf{x}$。这是一个线性高斯模型（Roweis和Gharamani，1999年）的例子，我们将在8.1.4中对其进行更广泛的研究。我们希望找到边缘概率$p(\mathbf{y})$，和条件概率$p(\mathbf{x} \mid \mathbf{y})$。这是一个在后续章节中经常出现的问题，它将证明得到一般的结果是很方便的。

我们将要使用到边缘和条件分布

$$p(\mathbf{x}) = \mathcal{N}(\mathbf{x} \mid \mathbf{\mu}, \mathbf{\Lambda}^{-1})$$

$$p(\mathbf{y} \mid \mathbf{x}) = \mathcal{N}(\mathbf{y} \mid \mathbf{Ax} + \mathbf{b}, \mathbf{L}^{-1})$$

这里$\mathbf{\mu}, \mathbf{A}$是均值的参数，$\mathbf{\Lambda}$和$\mathbf{L}$是精度矩阵。如果$\mathbf{x}$有$M$维，$\mathbf{y}$有$D$维，那么矩阵$\mathbf{A}$的大小为$D \times M$。

这里我们有一个在$\mathbf{x}$和$\mathbf{y}$上的联合分布的表达式。为了如此，我们定义

$$\mathbf{z} = \begin{pmatrix}
\mathbf{x} \\
\mathbf{y}
\end{pmatrix}$$

那么考虑一个对数联合分布

$$\begin{align*}\ln p(\mathbf{z})
&= \ln p(\mathbf{x}) + \ln p(\mathbf{y} \mid \mathbf{x}) \\
&= -\frac{1}{2}(\mathbf{x} - \mathbf{\mu})\mathbf{\Lambda}(\mathbf{x} - \mathbf{\mu}) - \frac{1}{2}(\mathbf{y} - \mathbf{Ax} - \mathbf{b})^T\mathbf{L}(\mathbf{y} - \mathbf{Ax} - \mathbf{b}) + \text{const}\end{align*}$$

其中$\text{const}$是独立于$\mathbf{x}$和$\mathbf{y}$的项。正如之前，我们看到这个是一个$\mathbf{z}$的二次型函数，因此$p(\mathbf{z})$也是一个高斯分布。为了找到高斯的精度，我们考虑2.102中的二次项，可以写作如下

$$-\frac{1}{2}\mathbf{x}^T(\mathbf{\Lambda} + \mathbf{A}^T\mathbf{L}\mathbf{A})\mathbf{x} - \frac{1}{2}\mathbf{y}^T\mathbf{L}\mathbf{y} + \frac{1}{2}\mathbf{y}^T\mathbf{LAx} + \frac{1}{2}\mathbf{x}^T\mathbf{A}^T\mathbf{Ly} = -\frac{1}{2}
\begin{pmatrix}
\mathbf{x} \\
\mathbf{y}
\end{pmatrix}^T
\begin{pmatrix}
\mathbf{\mathbf{\Lambda}} + \mathbf{A}^T\mathbf{L}\mathbf{A})\mathbf{x} & -\mathbf{A}^T\mathbf{L}\\
-\mathbf{LA} & \mathbf{L}
\end{pmatrix}
\begin{pmatrix}
\mathbf{x} \\
\mathbf{y}
\end{pmatrix} = -\frac{1}{2}\mathbf{z}^T\mathbf{R}\mathbf{z}$$

所以在$\mathbf{z}$上的高斯分布有精度矩阵（逆协方差）

$$\mathbf{R} = \begin{pmatrix}
\mathbf{\Lambda} + \mathbf{A}^T\mathbf{LA} & - \mathbf{A}^T\mathbf{L} \\
-\mathbf{LA} & \mathbf{L}
\end{pmatrix}$$

协方差矩阵就是去精度的逆，可以使用2.76中的形式来表示

$$\text{cov}[\mathbf{z}] = \mathbf{R}^{-1} = 
\begin{pmatrix}
\mathbf{\Lambda}^{-1} & \mathbf{\Lambda}^{-1}\mathbf{A}^T \\
\mathbf{A}\mathbf{\Lambda}^{-1} & \mathbf{L}^{-1} + \mathbf{A}\mathbf{\Lambda}^{-1}\mathbf{A}^T
\end{pmatrix}$$

类似地，我们也可以通过2.102中的线性项找到在$\mathbf{z}$上的高斯分布，如下给出

$$\mathbf{x}^T\mathbf{\Lambda}\mathbf{\mu} - \mathbf{x}^T\mathbf{A}^T\mathbf{Lb} + \mathbf{y}^T\mathbf{Lb} = 
\begin{pmatrix}
\mathbf{x} \\
\mathbf{y}
\end{pmatrix}
\begin{pmatrix}
\mathbf{\Lambda\mu} -  \mathbf{A}^T\mathbf{Lb}\\
\mathbf{Lb}
\end{pmatrix}$$

使用通过多元高斯完全二次型平方得出先前2.71的结果，我们得到如下$\mathbf{z}$的均值

$$\mathbb{E}[\mathbf{z}] = \mathbf{R}^{-1}
\begin{pmatrix}
\mathbf{\Lambda\mu} -  \mathbf{A}^T\mathbf{Lb}\\
\mathbf{Lb}
\end{pmatrix}$$

再使用2.105，然后我们得到

$$\mathbb{E}[\mathbf{z}] = 
\begin{pmatrix}
\mathbf{\mu} \\
\mathbf{A\mu} + \mathbf{b}
\end{pmatrix}$$

现在我们找一个在所有$\mathbf{x}$上的边缘分布$p(\mathbf{y})$的表达式。回想一下，当用分块协方差矩阵来表示时，高斯随机向量的一个子集上的边缘分布采用了一个特别简单的形式。特别是，它的均值和协方差如2.92和2.93给出的那样。使用2.105和2.108，我们看到边缘分布$p(\mathbf{y})$的均值和协方差如下给出

$$\mathbb{E}[\mathbf{y}] = \mathbf{A\mu} + \mathbf{b}$$

$$\text{cov}[\mathbf{y}] = \mathbf{L}^{-1} + \mathbf{A}\mathbf{\Lambda}^{-1}\mathbf{A}^T$$

这个结果的一个特别的例子，当$\mathbf{A} = \mathbf{I}$，退化为两个高斯的卷积，我们看到卷积的均值是两个高斯均值的和，卷积的写反差是它们协方差的和。

最终，我们找到条件$p(\mathbf{x} \mid \mathbf{y})$上的表达式。回想一下，条件分布的结果很容易使用分块精度矩阵来表达，使用2.73和2.75。将这些结果应用到2.105和2.108中，我们看到条件分布$p(\mathbf{x} \mid \mathbf{y})$的均值和协方差如下

$$\mathbb{E}[\mathbf{x} \mid \mathbf{y}] = (\mathbf{\Lambda} + \mathbf{A}^T\mathbf{LA})^{-1}\{\mathbf{A}^T\mathbf{L}(\mathbf{y} - \mathbf{B}) + \mathbf{\Lambda}\mathbf{\mu}\}$$

$$\text{cov}[\mathbf{x} \mid \mathbf{y}] = (\mathbf{\Lambda} + \mathbf{A}^T\mathbf{LA})^{-1}$$

这个条件的卷积可以看成是贝叶斯定理的例子。我们可以将分布$p(\mathbf{x})$解释为$\mathbf{x}$上的先验分布。如果变量$\mathbf{y}$被观测到，那么条件分布$p(\mathbf{x} \mid \mathbf{y})$代表的就是相应$\mathbf{x}$上的后验分布。有了边缘和条件分布，我们有效地表达联合分布$p(\mathbf{z}) = p(\mathbf{x})p(\mathbf{y} \mid \mathbf{x})$和形式$p(\mathbf{x} \mid \mathbf{y})p(\mathbf{y})$。这些结果可以总结如下

> **<font color='red'>边缘和条件高斯</font>**

> 给定一个$\mathbf{x}$上的边缘高斯和在给定$\mathbf{x}$下的$\mathbf{y}$的条件高斯分布

$$p(\mathbf{x}) = \mathcal{N}(\mathbf{x} \mid \mathbf{\mu}, \mathbf{\Lambda}^{-1})$$
$$p(\mathbf{y} \mid \mathbf{x}) = \mathcal{N}(\mathbf{y} \mid \mathbf{A\mu} + \mathbf{b}, \mathbf{L}^{-1})$$

> $\mathbf{y}$的边缘分布，和在给定$\mathbf{y}$时$\mathbf{x}$的条件分布如下给出

$$p(\mathbf{y}) = \mathcal{N}(\mathbf{y} \mid \mathbf{A\mu} + \mathbf{b}, \mathbf{L}^{-1} + \mathbf{A}\mathbf{\Lambda}^{-1}\mathbf{A}^T$$
$$p(\mathbf{x} \mid \mathbf{y}) = \mathcal{N}(\mathbf{x} \mid \mathbf{\Sigma}\{\mathbf{A}^TL(\mathbf{y} - \mathbf{b}) + \mathbf{\Lambda}\mu\}, \mathbf{\Lambda})$$

> 其中
$$\mathbf{\Sigma} = (\mathbf{\Lambda} + \mathbf{A}^T\mathbf{L}\mathbf{A})^{-1}$$

#### 2.3.4 高斯最大似然

给一个数据集$\mathbf{X} = (\mathbf{x}_1, \cdots, \mathbf{x}_N)^T$它的观测值$\{\mathbf{x}_n\}$假设是独立采样自多元高斯分布，我们可以使用最大似然来评估分布的参数。对数似然如下给出

$$\ln p(\mathbf{X} \mid \mathbf{\mu}, \mathbf{\Sigma}) = -\frac{ND}{2}\ln (2\pi) - \frac{N}{2}\ln |\mathbf{\Sigma}| - \frac{1}{2}\sum_{n=1}^N(\mathbf{x}_n - \mu)^T\mathbf{\Sigma}^{-1}(\mathbf{x}_n - \mu)$$

简单地重排下，我们看到似然函数依赖于数据集只有两项

$$\sum_{n-1}^N\mathbf{x}_n, \hspace{1.0cm} \sum_{n=1}^N\mathbf{x}_n\mathbf{x}_n^T$$

这写都是高斯分布的充分统计。使用附录2.19，表示$\mathbf{\mu}$的对数似然的导数如下

$$\frac{\partial}{\partial\mathbf{\mu}}\ln p(\mathbf{X} \mid \mathbf{\mu}, \mathbf{\Sigma}) = \sum_{n=1}^N\mathbf{\Sigma}^{-1}(\mathbf{x}_n - \mathbf{\mu})$$

令其导数为0，我们获得最大似然评估解的均值如下

$$\mathbf{\mu}_{\text{ML}} = \frac{1}{N}\sum_{n=1}^N\mathbf{x}_n$$

这是观测到的数据点集的均值。关于$\mathbf{\Sigma}$的最大化则更为复杂。最简单的方法就是忽略对称约束，证明得到的结果是堆成的。这一结果的替代推导，明确限定了对称性和正定性约束，这可在Magnus和Neudecker (1999年)中找到。结果如期望的，取如下形式

$$\mathbf{\Sigma}_{\text{ML}} = \frac{1}{N}\sum_{n=1}^N(\mathbf{\mu}_n - \mathbf{\mu}_{\text{ML}})(\mathbf{\mu}_n - \mathbf{\mu}_{\text{ML}})^T$$

这里包含$\mathbf{\mu}_{\text{ML}}$，是因为这个结果是联合最大化表示的$\mathbf{\mu}$和$\mathbf{\Sigma}$。需要注意的是对于$\mathbf{\mu}_{\text{ML}}$的解不依赖于$\mathbf{\Sigma}_{\text{ML}}$，所以我们可以先评估$\mathbf{\mu}_{\text{ML}}$，然后使用这个区评估$\mathbf{\Sigma}_{\text{ML}}$。

如果我们在真实分布下，评估最大似然解的期望，我们会得到如下解

$$\mathbb{E}[\mathbf{\mu}_{\text{ML}}] = \mathbf{\mu}$$

$$\mathbb{E}[\mathbf{\Sigma}_{\text{ML}}] = \frac{N-1}{N}\mathbf{\Sigma}$$

我们看到均值最大似然评估的期望等于真实均值。而，最大似然评估的方差它的期望是小于真实值，因此它存在偏差。我们可以通过定义一个不同的评估器$\tilde{\mathbf{\Sigma}}$来纠正偏差如下

$$\tilde{\mathbf{\Sigma}} = \frac{1}{N-1}(\mathbf{\mu}_n - \mathbf{\mu}_{\text{ML}})(\mathbf{\mu}_n - \mathbf{\mu}_{\text{ML}})^T$$

从2.122和2.124中清楚地可以看到，$\tilde{\mathbf{\Sigma}}$的期望等于$\mathbf{\Sigma}$。

#### 2.3.5 序列评估

我们讨论的高斯分布参数的最大似然解，提供了一个更便利的机会，可以给出更多的一般的关于最大似然序列评估的主题的讨论。序列方法一次只处理一部分数据，然后丢掉，它们对于在线应用很重要，同时会有大量的数据会加进来，所以对所有数据的批处理流程就更加灵活。

考虑结果2.121均值$\mathbf{\mu}_{\text{ML}}$的最大似然评估，当它依赖于$N$个观测值，我们表示成$\mathbf{\mu}_{\text{ML}}^{N}$。如果我们从最早数据点$\mathbf{x}_N$中进行分析，我们得到

$$\begin{align*}
\mathbf{\mu}_{\text{ML}}^{N}
&= \frac{1}{N}\sum_{n=1}^N\mathbf{x}_n \\
&= \frac{1}{N}\mathbf{x}_N + \frac{1}{N}\sum_{n=1}^{N-1}\mathbf{x}_N \\
&= \frac{1}{N}\mathbf{x}_N + \frac{N-1}{N}\mathbf{\mu}_{\text{ML}}^{(N-1)} \\
&= \mathbf{\mu}_{\text{ML}}^{(N-1)} + \frac{1}{N}\bigr(\mathbf{x}_N - \mathbf{\mu}_{\text{ML}}^{(N-1)}\bigr)
\end{align*}$$

这个结果也很好解释，如下。在观测$N-1$数据点后，我们使用$\mathbf{\mu}_{\text{ML}}^{(N-1)}$来评估$\mathbf{\mu}$。我们现在观测数据点$\mathbf{x}_N$，我们通过移除掉老的一小部分的数据得到修正的评估$\mathbf{\mu}_{\text{ML}}^{(N)}$，它与$1/N$成比例，修正的方向是$\bigr(\mathbf{x}_N - \mathbf{\mu}_{\text{ML}}^{(N-1)}\bigr)$。需要注意的是，随着$N$的增加，从成功数据点的贡献将变得越来越小。

2.126的结果将如1.121的批处理结果一样给出同样的答案，因为这两种形式是等价。但是，我们并不能总是使用这样的规则得到序列化算法，所以我们需要寻找一个更加通用的序列化学习的形式，这将我们引导到Robbins-Monro算法。考虑一对随机变量$\theta$和$z$表示成联合分布$p(z, \theta)$。在给定$\theta$下$z$的条件期望，定义成确定性函数$f(\theta)$如下给出

$$f(\theta) \equiv \mathbb{E}[z \mid \theta] = \int zp(z \mid \theta)dz$$

如图2.10中示意图所示。以这种方法定义的函数称作回归函数。



<center>图2.10 两个相关随机变量$z$和$\theta$的示意图，它的回归函数$f(\theta)$由条件期望$\mathbb{E}[z \mid \theta]$条件分布给出。Robbins-
Monro算法提供了一个通用的序列化流程来找到这种函数的根$\theta^*$。</center>

我们目标是找到的根$\theta^*$是在$f(\theta^*) = 0$。如果我们有一个大的观测数据集$z$和$\theta$，那么我们嗯呢各个直接对回归函数建模，然后得到根的评估。假设，我们一次观测到一个$z$的值，我们希望找到一个$\theta^*$序列化评估方案。下面的通用流程是用来解决这个问题的，这是由Robbins和Monro (1951年)给出的。我们将假设条件变量$z$是有限的，有

$$\mathbb{E}[(z - f)^2 \mid \theta] < ∞$$

我们也将，不失一般性，考虑对于$\theta > \theta^*$，有$f(\theta) > 0$，而$\theta < \theta^*$，有$f(\theta) < 0$，如图2.10所示的那样。Robbins-Monro流程是定义了一个成功评估根$\theta^*$的序列，如下

$$\theta^{(N)} = \theta^{(N-1)} + a_{N-1}z(\theta^{(N-1)})$$

其中$z(\theta^{(N)})$是在$\theta$值取$\theta^{(N)}$的时候，观测到的$z$值。系数$\{a_N\}$表示的是正数的序列化，满足如下条件

$$\lim_{N \to ∞}a_N = 0$$

$$\sum_{N=1}^∞a_N = ∞$$

$$\sum_{N=1}^∞a_N^2 < ∞$$

2.129给出的序列化评估有一定概率会收敛到一根，这是可以证明的（Robbins和Monro; Fukunaga, 1990年）。需要注意的是，2.130的第一个条件确保了成功修正是保证幅度的减小，使得最终可以收敛到一个极值。2.131的第二个条件需要确保算法不会再短期内收敛到根，而2.132的第三个条件是需要确保累计的噪声有一个有限的方差，不会破坏收敛性。

现在让我们考虑使用Robbins-Monro算法一个通用的最大化似然问题是怎样被解决的。通过定义，最大化似然解$\theta^{\text{ML}}$是在固定点的对数似然函数，因此满足

$$\frac{\partial}{\partial \theta}\left\{\frac{1}{N}\sum_{n=1}^N\ln p(\mathbf{x}_n \mid \theta)\right\} \Bigg|_ {\theta_{\text{ML}}} = 0$$

交换求导和求和，取极限$N \to ∞$，我们有

$$\lim_{N \to ∞}\frac{1}{N}\sum_{n=1}^N\frac{\partial}{\partial \theta} \ln p(x_n \mid \theta) = \mathbb{E}_x \left[\frac{\partial}{\partial \theta}\ln p(x \mid \theta)\right]$$

所以我们看到找到UI打死人解相应的就是找回归函数的根。我们因此可以应用Robbins-Monro流程，现在取如下形式

$$\theta^{(N)} = \theta^{(N-1)} + a_{N-1}\ln p(x_N \mid \theta^{(N-1)})$$

作为一个特别的资历，我们再次考虑高斯分布的序列化评估，参数$\theta^{(N)}$的评估$\mu_{\text{ML}}^{(N)}$为高斯的均值，随机变量$z$如下给出

$$z = \frac{\partial}{\partial \mu_{\text{ML}}} \ln p(x \mid \mu_{\text{ML}}, \sigma^2) = \frac{1}{\sigma^2}(x - \mu_{\text{ML}})$$

因此$z$的分布是均值为$\mu - \mu_{\text{ML}}$的高斯，如图2.11所示。将2.136代入2.135中，我们得到2.126的单变量形式，提供我们选择的系数$a_N$可以有形式$a_N = \sigma^2 / N$。需要注意的是尽管我们是聚焦于单变量，同样的技术，应用到多变量例子中（Blum, 1965年），也会在系数$a_N$上有同样的限制2.130-2.132。



<center>图2.11 在高斯分布中，$\theta$相应的参数是均值$\mu$，如图2.10中的回归函数取一段直线，如红色所示。在这个例子汇总，随机变量$z$对应的对数似然函数的倒数，是由$(x - \mu_{\text{ML}})/\sigma^2$给出的，它的期望定义额回归函数是一条直线由$(\mu - \mu_{\text{ML}})/\sigma^2$。回顾函数的根对应的最大似然函数评估$\mu_{\text{ML}}$。</center>

#### 2.3.6 高斯贝叶斯推断

最大似然框架给出了参数$\mathbf{\mu}$和$\mathbf{\Sigma}$的点估计。现在我们通过介绍在这些参数上的先验分布来研究贝叶斯处理。让我们先来考虑一个单高斯随机变量$x$的简单例子。我们假设方差$\sigma^2$是已知的，我们考虑在给出$N$个观测值$\mathbf{X} = \{x_1,\cdots,x_N\}$的集合的均值$\mu$推断的任务。似然函数，那是给出$\mu$的观测数据的概率，被视作是$\mu$的函数，如下给出

$$p(\mathbf{X} \mid \mathbf{\mu}) = \prod_{n=1}^Np(x_n \mid \mu) = \frac{1}{(2\pi\sigma^2)^{N/2}}\text{exp}\left\{-\frac{1}{2\sigma^2}\sum_{n=1}^N(x_n - \mu)^2\right\}$$

我们再次强调似然函数$p(\mathbf{X} \mid \mu)$不是在$\mu$上的概率分布，也没有标准化。（注：这里$\mu$是已知的，只是对于不同的$x$，会有不同的概率，它是对于$x$的概率。）

我们看到似然函数取了有$\mu$二次型的指数项。因此如果我们选择一个高斯的先验，它将是似然函数的共轭分布，因为它相应的后验将是两个二次函数$\mu$额指数乘积，因此，他也是一个高斯。我们因而可以取先验为

$$p(\mu) = \mathcal{N}(\mu \mid \mu_0, \sigma_0^2)$$

后验分布给出如下

$$p(\mu \mid \mathbf{X}) \propto p(\mathbf{X} \mid \mu)p(\mu)$$

涉及指数中的完全平方的简单操作证明，后验分布给出如下

$$p(\mu \mid \mathbf{X}) = \mathcal{N}(\mu \mid \mu_N, \sigma_N^2)$$

其中

$$\mu_N = \frac{\sigma^2}{N\sigma_0^2 + \sigma^2}\mu_0 + \frac{N\sigma_0^2}{N\sigma_0^2 + \sigma^2}\mu_{\text{ML}}$$

$$\frac{1}{\sigma_N^2} = \frac{1}{\sigma_0^2} + \frac{N}{\sigma^2}$$

这里的$\mu_{\text{ML}}$是最大似然解$\mu$给出的样本均值

$$\mu_{\text{ML}} = \frac{1}{N}\sum_{n=1}^Nx_n$$

花一些时间来研究后验均值和方差的形式是有价值的。首先，我们注意到后验分布的均值如2.141给出的，它是由先验均值$\mu_0$和最大似然解$\mu_{\text{ML}}$组成。如果观测数据点的数量$N=0$，那么2.141就会退为正如所期望的先验均值。对于$N \to ∞$，后验均值就是由最大似然解给出的。类似地，考虑2.142后验分布方差的结果。我们看到这很大部分都是自然地使用方差的倒数来表示，这叫做精度。此外，精度是相加的，这个后验的数据精度是由先验数据精度的来自每个观测数据点的一个贡献叠加。随着我们增加观测数据点的数量，精度也会极具增加，相应的后验分布也会极具减少方差。没有观测到数据点时，我们有先验方差，而如果数据点的数量$N \to ∞$，方差$\sigma_N^2$趋于0，后验分布在最大似然解附件变得很陡峭。因此，我们可以看到2.143给出的$\mu$的点估计的最大似然结果是恰好从无限数据量的贝叶斯形式主义中恢复过来。需要注意的是对于有限$N$，如果我们取极限$\sigma_0^2 \to ∞$，先验有无限方差，那么后验2.141蜕变为最大似然结果，从2.142中的后验方差则由$\sigma_N^2 = \sigma^2 / N$给出。

图2.12画出的是贝叶斯推断分析高斯分布均值。将这个结果推广到$D$维高斯随机变量$\mathbf{x}$，已知方差，未知均值是很直接的。



<center>图2.12 画的是高斯分布均值$\mu$的贝叶斯推断，这里假设方差是已知的。曲线显示的是在$\mu$上的先验（曲线标签$N=0$），它本身也是高斯，随着$N$数据点的增长给出后验分布2.140。数据点是从均值0.8，方差0.1的高斯中生成的，先验的选择有均值为0.同时，先验和似然函数，方差被设置成真实值。</center>

我们已经看到最大似然高斯平均值的表达式可以重新转换成顺序更新公式，其中$N$个数据点的均值是观察$N-1$个数据点均值和$\mathbf{x}_N$个数据点的共同贡献的结果。事实上，贝叶斯范式很自然地导向了序列化视角的推断问题。为了在高斯均值推断内容中看到这个，我们写下最终$\mathbf{x}_N$个数据点对后验分布的贡献的分离式子

$$p(\mathbf{\mu} \mid D) \propto \left[p(\mathbf{\mu})\prod_{n=1}^{N-1}p(\mathbf{x}_n \mid \mathbf{\mu})\right]p(\mathbf{x}_N \mid \mathbf{\mu})$$

方括号中的项（归一化系数）仅是观测$N-1$个数据点的后验分布。我们看到这可以视作是一个先验分布，在结合使用贝叶斯定理的似然函数关联的数据点$\mathbf{x}_N$，在观测到$N$个数据点后，得到后验分布。贝叶斯推断的序列化观点非常通用，应用到任何问题时，这些观测数据都假设是独立同分布的。

迄今为止，我们已经假设数据上的高斯分布的方差是已知的，我们的目标就是推断均值。现在让我们假设均值是已知的，我们希望推断方差，同时，如果我们选择共轭形式的先验分布，这个计算将相当简单。结果使用精度$\lambda \equiv 1/\sigma^2$会相当方便。$\lambda$的似然函数取如下

$$p(\mathbf{X} \mid \lambda) = \prod_{n=1}^N\mathcal{N}(x_n \mu, \lambda^{-1})\propto\lambda^{N/2}\text{exp}\left\{-\frac{\lambda}{2}\sum_{n=1}^N(x_n - \mu)^2\right\}$$

相应的共轭先验应该成比例于$\lambda$的幂和$\lambda$的线性函数的指数。相应的gamma分布定义如下

$$\text{Gam}(\lambda \mid a, b) = \frac{1}{\Gamma(a)}b^a\lambda^{a-1}\text{exp}(-b\lambda)$$

这里$\Gamma(a)$是1.141中定义的gamma函数，它确保是被正确标准化过的。如果$a > 0$，gamma分布有一个有限积分，如果$a \ge 1$，分布本身是有限的。图2.13画出的各种$a$和$b$值。gamma分布的均值和方差如下给出

$$\mathbb{E}[\lambda] = \frac{a}{b}$$

$$\text{var}[\lambda] = \frac{a}{b^2}$$

> 注：$\Gamma(a) \equiv \int_0^∞u^{a-1}e^{-u}du$



<center>图2.13 图中画的是2.146定的是各种参数$a$和$b$的值的gamma分布。</center>

考虑一个先验分布$\text{Gam}(\lambda \mid a_0, b_0)$。如果我们与2.145的似然相乘，那么我们得到后验分布

$$p(\lambda \mid \mathbf{X}) \propto \lambda^{a_0 - 1}\lambda^{N/2}\text{exp}\left\{-b_0\lambda - \frac{\lambda}{2}\sum_{n=1}^N(x_n - \mu)^2\right\}$$

这里我们看到它也是一个gamma分布$\text{Gam}(\lambda \mid a_N, b_N)$，其中

$$a_N = a_0 + \frac{N}{2}$$

$$b_N = b_0 + \frac{1}{2}\sum_{n=1}^N(x_n - \mu)^2 = b_0 + \frac{N}{2}\sigma_{\text{ML}}^2$$

其中$\sigma_{\text{ML}}^2$是最大似然的方差。需要注意的是2.149中没必要跟踪先验中的标准化常量和似然函数，是因为，如果需要的，正确系数可以在节尾使用2.146的gamma分布的标准化形式。（注：这里说的是在节尾时，加上除以$\Gamma(a)$函数。）

从2.150中，我们看到观测$N$个数据点的效果是增加系数$a$的值乘以$N/2$。因此我们可以解释先验中的参数$a_0$，使用先验观测的$2a_0$项更有效。类似地，从2.151中我们可以看到$N$个数据点贡献$N\sigma_{\text{ML}}^2/2$给参数$b$，其中$\sigma_{\text{ML}}^2$是方差，所以我们可以解释先验参数$b_0$，产生于$2a_0$有效先验观测，具有方差$2b_0/2a_0 = b_0/a_0$。回想一下，我们对Dirichlet先验做了类似的解释。这些分布都是指数家族的例子，我们应该看到一个共轭先验在有效数据数据句点上式一个通用的指数家族分布。

代替使用精度，我们可以考虑使用方差。在本例中的共轭先验称为逆gamma分布，虽然我们不会进一步论这个问题，因为我们发现使用精度更方便。

现在假设讨均值和精度都是未知的。为了找到一个共轭先验，我们考虑似然函数在$\mu$和$\lambda$上的相关性

$$p(\mathbf{X} \mid \mu, \lambda) = \prod_{n=1}^N(\frac{\lambda}{2\pi})^{1/2}\text{exp}\left\{-\frac{\lambda}{2}(x_n - \mu)^2\right\} \propto \left[\lambda^{1/2}\text{exp}\left(-\frac{\lambda\mu^2}{2}\right)^N\text{exp}\left\{\lambda\mu\sum_{n=1}^Nx_n - \frac{\lambda}{2}\sum_{n=1}^Nx_n^2\right\}\right]$$

我们现在希望先验分布$p(\mu, \lambda)$在$\mu$和$\lambda$上有同样的功能性相关的似然函数，它因此可以取如下形式

$$p(\mu, \lambda) \varpropto \left[\lambda^{1/2}\text{exp}\left(-\frac{\lambda\mu^2}{2}\right)\right]^{\beta}\text{exp}\left\{c\lambda\mu - d\lambda\right\} = \text{exp}\left\{-\frac{\beta\lambda}{2}(\mu-c\beta)^2\right\}\lambda^{\beta/2}\text{exp}\left\{-\left(d-\frac{c^2}{2\beta}\right)\lambda\right\}$$

其中$c, d$和$\beta$是常量。因为我们总有$p(\mu, \lambda) = p(\mu \mid \lambda)p(\lambda)$，我们可以通过检查找到$p(\mu \mid \lambda)p(\lambda)$。特别的，我们看到$p(\mu \mid \lambda)$是一个高斯，它的精度为$\lambda$的线性函数，$p(\lambda)$是gamma分布，所以标准化先验取如下形式

$$p(\mu, \lambda) = \mathcal{N}(\mu \mid \mu_0, (\beta\lambda)^{-1})\text{Gam}(\lambda \mid a, b)$$

其中我们定义了新的常量$\mu_0 = c / \beta$，$a = 1 + \beta/2$，$b = d - c^2/2\beta$。2.154的分布称作正太gamma或高斯gamma分布，如图2.14所示。需要注意的是这并不是简单的在参数$\mu$上的高斯先验和参数$\lambda$上的gamma先验的相乘。即使我们选择的先验$\mu$和$\lambda$是相互独立的，后验分布也会呈现出一种在$\mu$的精度和$\lambda$值的耦合。



<center>图2.14 对于参数$\mu_0$，$\beta=2, a=5$和$b=6$的正太gamma分布（2.154）。</center>

在$D$维变量$\mathbf{x}$的多元高斯分布$\mathbf{N(\mathbf{x} \mid \mathbf{\mu}, \mathbf{\Lambda}^{-1})}$的例子中，共轭先验分布的均值$\mathbf{\mu}$，假设精度是已知的，仍然是一个高斯。对于已知均值和未知精度矩阵$\mathbf{\Lambda}$，共轭先验是Wishart分布如下

$$\mathcal{W}(\mathbf{\Lambda} \mid \mathbf{W}, \mathcal{v}) = B \mid \mathbf{\Lambda} \mid^{(\mathcal{v} - D - 1)/2}\text{exp}\left(-\frac{1}{2}\text{Tr}(\mathbf{W}^{-1})\mathbf{\Lambda}\right)$$

其中$\mathcal{v}$被称为分布的自由度数量，$\mathbf{W}$是一个$D \times D$的比例矩阵，$\text{Tr}(\cdot)$表示的是迹。标准化常量$B$如下

$$B(\mathcal{W}, \mathcal{v}) = |\mathcal{W}|^{-\mathcal{v}/2}\left(2^{\mathcal{v}D/2}\pi^{D(D-1)/4}\prod_{i=1}^D\mathbf{\Gamma}\left(\frac{\mathcal{v} + 1 - i}{2}\right)\right)^{-1}$$

同样，我们也可以在协方差矩阵本身定义一个先验，而不是在精度矩阵上，这会有产生一个逆Wishart分布，虽然我们不会进一步讨论这个问题。如果均值和精度是未知的，那么，遵循与单变量例子类似的推理方式，共轭先验如下

$$p(\mathbf{\mu}, \mathbf{\Lambda} \mid \mathbf{\mu}_0, \beta, \mathbf{W}, \mathcal{v}) = \mathcal{N}(\mathbf{\mu}\mid\mathbf{\mu}_0,(\beta\mathbf{\Lambda})^{-1})\mathcal{W}(\mathbf{\Lambda} \mid \mathbf{W}, \mathcal{v})$$

这被称为正太Wishart，或高斯Wishart分布。

#### 2.3.7 学生t分布

我们看到高斯精度的共轭先验是由gamma分布给出的。如果我们有一个单变量高斯$\mathcal{N}(x \mid \mu, \tau^{-1})$，还有一个Gamma先验$\text{Gam}(\tau \mid a,b)$，同时我们对精度精度，我们得到边缘分布$x$的形式如下

$$\begin{align*}
p(x \mid \mu, a, b) 
&= \int_0^{∞}\mathcal{N}(x \mid \mu, \tau^{-1})\text{Gam}(\tau \mid a, b)d\tau \\
&= \int_0^{∞}\frac{b^ae^{(-b\tau)}\tau^{a-1}}{\Gamma(a)}\left(\frac{\tau}{2\pi}\right)^{1/2}\text{exp}\left\{-\frac{\tau}{2}(x-\mu)^2\right\}d\tau \\
&= \frac{b^a}{\Gamma(a)}\left(\frac{1}{2\pi}\right)^{1/2}\left[b + \frac{(x-\mu)^2}{2}\right]^{-a-1/2}\Gamma(a+1/2)
\end{align*}$$

其中我们替换一下变量$z=\tau[b+(x-\mu)^2/2]$。为了方便，我们定义新参数由$\mathcal{v}=2a$和$\lambda=a/b$给出，就此，我们有分布$p(x \mid \mu, a, b)$取如下形式

$$\text{St}(x \mid \mu, \lambda, \mathcal{v}) = \frac{\Gamma(\mathcal{v}/2 + 1/2)}{\Gamma(\mathcal{v}/2)}\left(\frac{\lambda}{\pi\mathcal{v}}\right)(\frac{\lambda}{\pi\mathcal{v}})\left[1 + \frac{\lambda(x-\mu)^2}{\mathcal{v}}\right]^{-\mathcal{v}/2-1/2}$$

这被称为学生t分布。参数$\lambda$优势被称作t分布的精度，一般来说它也不等于方差的倒数。参数$\mathcal{v}$被称为自由度，它的有效性如图2.15所示。对于$\mathcal{v}=1$的特殊例子，t分布退为柯西分布，当$\mathcal{v} \to ∞$，t分布$\text{St}(x \mid \mu, \lambda, \mathcal{v})$变成带有均值$\mu$和精度$\lambda$的高斯$\mathcal{N}(x \mid \mu, \lambda^{-1})$。



<center>图2.15 对于$\mu=0$和$\lambda=1$的各种$\mathcal{v}$值的学生t分布（2.159），极限$\mathcal{v} \to ∞$表示的是相应的均值$\mu$和精度$\lambda$的高斯分布。</center>

从2.158中，我们看到学生t分布是通过一个有着一样的均值，不同的精度的无限量的高斯分布的相加。这可以解释为一个无限混合高斯（混合高斯将在2.3.9节中详细讨论）。结果是一个有着比高斯更长尾部的分布，如2.15所示。这给出了t分布一个重要的特性称为鲁棒性，这意味着它比高斯的敏感度更低，使得能够发现一些异常的数据点。t分布的铝棒性如图2.16所示，比较了高斯和t分布的最大似然。需要注意的是t分布的最大似然解可以使用EM算法来得到。这里我们看到少量的异常点，对于t分布来说，要比高斯所具有的意义更少。在实际应用中，异常点的产生要么是因为生成数据的流程，先用的分布有着一个厚重的长尾，要么是因为是标记错误。铝棒性对于回归问题来说依然是很重要的特性。一点也不惊讶的是，对于回归问题的最小二乘表现的也并不鲁邦，因为它是在（条件）高斯分布下的最大似然。如果是基于一个厚重长尾分布的回归模型，比如t分布，我们会得到更加鲁邦的模型。

如果我们回到2.158中，带入相应的参数$\mathcal{v}=2a, \lambda=a/b$，和$\eta=\tau b/a$，我们看到t分布可以写作如下形式

$$\text{St}(x \mid \mu, \Lambda, \mathcal{v}) = \int_0^∞\mathcal{N}(x \mid \mu, (\eta\lambda)^{-1})\text{Gam}(\eta \mid \mathcal{v}/2, \mathcal{v}/2)d\eta$$

我们可以将多元高斯$\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}\mathbf{\Lambda})$进行推广，得到相应的多元学生t分布形式

$$\text{St}(\mathbf{x} \mid \mathbf{\mu}, \mathbf{\Lambda}, \mathcal{v}) = \int_0^∞\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}, (\eta\mathbf{\Lambda})^{-1})\text{Gam}(\eta \mid \mathcal{v}/2, \mathcal{v}/2)d\eta$$

如单变量的例子，使用同样的技术，我们进行整合如下

$$\text{St}(\mathbf{x} \mid \mathbf{\mu}, \mathbf{\Lambda}, \mathcal{v}) = \frac{\Gamma(D/2+\mathcal{v}/2)}{\Gamma(\mathcal{v}/2)}\frac{|\mathbf{\Lambda}|^{1/2}}{(\pi\mathcal{v})^{D/2}}\left[1 + \frac{\Delta^2}{\mathcal{v}}\right]^{-D/2-\mathcal{v}/2}$$

其中$D$是$\mathbf{x}$的维度，$\Delta$是平方马氏距离定义如下

$$\Delta^2 = (\mathbf{x} - \mathbf{\mu})^T\mathbf{\Lambda}(\mathbf{x} - \mathbf{\mu})$$

这是学生t分布的多元形式，满足如下性质

$$\mathbb{E}[\mathbf{x}] = \mathbf{\mu}, \hspace{0.5cm} \text{if} \hspace{0.2cm} \mathcal{v} > 1$$

$$\text{cov}[\mathbf{x}] = \frac{\mathcal{v}}{(\mathcal{v}-2)}\mathbf{\Lambda}^{-1}, \hspace{0.5cm} \text{if} \hspace{0.2cm} \mathcal{v} > 2$$

$$\text{model}[\mathbf{x}] = \mathbf{\mu}$$

对于单变量来说结果也是如此。

#### 2.3.8 周期变量

尽管高斯分布有着很大的实际意义，在它本身的领域，和对于复杂概率模型的构建，会有一些情形，是它们作为连续变量的密度模型而不能胜任的。在实际应用中的一个重要的例子，就是周期变量。

一个周期变量的例子就是在一个特定地理位置的风向。我们可能，比如，测量了数天的风向，希望使用一个参数化分布来概括它。另一个例子是阳历时间，这里我们可能会对量化建模，相信它是24小时，或者一个季度为周期的。这种量化可以很方便地使用极坐标系$0 \le \theta \le 2\pi$来表示。

我们可能想选择某些方向作为开始来处理周期变量，然后应用一个常规分布，比如高斯。这种方法，给出的结果是与起始的随意选择强相关的。假设，比例，我们有两个观测值$\theta_1 = 1°$和$\theta_2 = 359°$，我们使用一个标准化的单变量高斯来对它进行建模。如果我们选择$0°$为开始，那么这个数据集的样本均值将会是$180°$，有$179°$的标准差，倘若我们选择$180°$为起始，那么均值会是$0°$，标准差为$1°$。我们亟待开发出一个用于处理周期变量的特别方法。

让我们考虑一个观测值数据集$\mathcal{D}=\{\theta_1,\cdots,\theta_N\}$的周期变量的均值评估问题。从现在开始，我们假设$\theta$是以弧度为测量单位。我们已经看到简单的平均$(\theta_1+\cdots+\theta_N)/N$将会是强协作相关。为了找到单变量均值的度量，我们注意到观测值可以被视作在单位圆上的数据点，可以被一个二维单元向量$\mathbf{x}_1,\cdots,\mathbf{x}_N$，其中$\|\mathbf{x}_n\|=1$来描述，如图2.17所示的那样。我们取向量$\{\mathbf{x}_n\}$的平均值如下

$$\bar{\mathbf{x}}=\frac{1}{N}\sum_{n=1}^N\mathbf{x}_n$$

可以找到对应的角$\hat{\theta}$的均值。非常明确的是，这个定义确保了均值的位置是与极坐标的起始是不相关。需要注意的是$\hat{\mathbf{x}}$一般地都是位于单位圆内的。笛卡尔坐标的观测值由$\bar{\mathbf{x}_n}=(\cos\theta_n, \sin\theta_n)$给出，我们可以写出笛卡尔坐标的均值，形式如$\bar{\mathbf{x}_n}=(\bar{r}\cos\bar{\theta}_n, \bar{r}\sin\bar{\theta}_n)$。带入到2.167中，等价的$x_1$和$x_2$组合，给出如下

$$\bar{r}\cos\bar{\theta}_n = \frac{1}{N}\sum_{n=1}^N\cos\theta_n, \hspace{0.5cm} \bar{r}\sin\bar{\theta}_n=\frac{1}{N}\sum_{n=1}^N\sin\theta_n$$

对其取比率，使用等式$\tan\theta = \sin\theta/\cos\theta$，我们可以解出$\bar{\theta}$如下

$$\bar{\theta} = \tan^{-1}\left\{\frac{\sum_n\sin\theta_n}{\sum_n\cos\theta_n}\right\}$$

一会儿，我们会看到这个结果是如何自然地产生自周期变量上的合适定义分布的最大似然
评估的。



<center>图2.17 图中表示的值$\theta_n$的在单位圆内的二维向量$\mathbf{x}_n$的周期变量。同时显示了这个变量的均值$\bar{\mathbf{x}}$。</center>

我们现在考虑高斯的周期推广，称为Mises分布。这里我们将注意力限制在单变量分布上，然而我们可以再任意维度的超球面上找到周期分布。对于周期分布的扩展性讨论，可以看Mardia和Jupp（2000年）。

为了方便，我们会考虑分布$p(\theta)$有着$2\pi$周期。定义在$\theta$上的任何概率分布$p(\theta)$必须不仅要是非负的，而且其积分要为1，但也要是周期性的。因此$p(\theta)$必须满足以下三个条件

$$p(\theta) \ge 0$$

$$\int_0^{2\pi}p(\theta)d\theta = 1$$

$$p(\theta + 2\pi) = p(\theta)$$

从2.172中，它遵循对于任意整数$M$，都有$p(\theta + M\pi) = p(\theta)$

我们很容易就可以找到类高斯分布可以满足这三个性质。考虑一个在二元变量$\mathbf{x}=(x_1, x_2)$的高斯分布，有均值$\mathbf{\mu}=(\mu_1, \mu_2)$，协方差矩阵$\mathbf{\Sigma}=\sigma^2\mathbf{I}$
，其中$\mathbf{I}$是一个$2 \times 2$的单位矩阵，所以

$$p(x_1, x_2) = \frac{1}{2\pi\sigma^2}\text{exp}\left\{-\frac{(x_1-\mu_1)^2 + (x_2-\mu_2)^2}{2\sigma^2}\right\}$$

常量$p(\mathbf{x})$等高线都是环，如图2.18所示。现在假设我们考虑沿着固定半径的圆的分布值。然后重构后的这个分布僵尸周期性的，尽管它并没有标准化。我们可以通过从笛卡尔坐标系$(x_1, x_2)$转换到极坐标系$(r,\theta)$来定义这个分布的形式，所以有

$$x_1 = r\cos\theta, \hspace{0.5cm} x_2=r\sin\theta$$

我们可以将均值$\mathbf{\mu}$映射到极坐标系中，写作如下

$$\mu_1=r_0\cos\theta_0, \hspace{0.5cm} \mu_2=r_0\sin\theta_0$$

接下来我们带入这些转换到二维高斯分布2.173中，然后定义条件单位圆$r=1$，需要注意的是我们感兴趣的仅是与$\theta$相关的。聚焦于高斯分布中的指数项，我们有

$$\begin{align*}
-&\frac{1}{2\sigma^2}\{(r\cos\theta - r_0\cos\theta_0)^2 + (r\sin\theta - r_0\sin\theta_0)^2\} \\
&= -\frac{1}{2\sigma^2}\{1+r_0^2 - 2r_0\cos\theta\cos\theta_0 - 2r_0\sin\theta\sin\theta_0\} \\
&= \frac{r_0}{\sigma^2}\cos(\theta-\theta_0) + \text{const}
\end{align*}$$

其中，'$\text{const}$'表示与$\theta$的i独立项，我么使用下面的三角恒等式

$$\cos^2A + \sin^2A = 1$$

$$\cos A\cos B + \sin A \sin B = \cos(A-B)$$

如果我们现在定义$m = r_0/\sigma^2$，我们得到沿着单位圆$r=1$的$p(\theta)$分布的最终表达式如下形式

$$p(\theta \mid \theta_0, m) = \frac{1}{2\pi I_0(m)}\text{exp}{m\cos(\theta - \theta_0)}$$

这称做von Mises分布，或circular正太。这里参数$\theta_0$是相应的分布均值，而$m$，被称为浓度参数，类似于高斯的逆方差（精度）。在2.179中的标准化系数表示成$I_0(m)$，这是一种零阶贝塞尔函数（Abramowitz和Stegun，1965年），定义如下

$$I_0(m) = \frac{1}{2\pi}\int_0^{2\pi}\text{exp}\{m\cos\theta\}d\theta$$

对于大$m$，分布近似高斯。von Mises分布如图2.19所示，函数$I_0(m)$如图2.20所示。



<center>图 2.20 贝尔赛函数$I_0(m)$如图2.180所示，同时带着的函数$A(m)$如2.186定义的。</center>

现在考虑von Mises分布的参数$\theta_0$和$m$的最大似然评估。对数似然函数如下给出

$$\ln p(\mathcal{D} \mid \theta_0, m) = -N\ln(2\pi) - N\ln I_0(m) + m\sum_{n=1}^N\cos(\theta_n - \theta_0)$$

对$\theta_0$求导，令其为0，则有

$$\sum_{n=1}^N\sin(\theta_n - \theta_0)=0$$

为了求解$\theta_0$，我么使用三角恒等式

$$\sin(A-B)=\cos B\sin A - \cos A \sin B$$

继而我们有

$$\theta_{\text{ML}} = \tan^{-1}\left\{\frac{\sum_n\sin\theta_n}{\sum_n\cos\theta_n}\right\}$$

我们很容易就认出这个结果是2.169早些我们得到的二维笛卡尔空间的观测均值。

类似地，最大化2.181来表示$m$，使用$I'_0(m)=I_1(m)$（Abramowitz和Stegun, 1965年），我们有

$$A(m)=\frac{1}{N}\sum_{n=1}^N\cos(\theta_n-\theta_0^{\text{ML}})$$

这里我们带入了最大化似然解$\theta_0^{\text{ML}}$（回想一下，这里我们执行了在$\theta$和$m$上的联合优化），我们定义

$$A(m) = \frac{I_1(m)}{I_0(m)}$$

函数$A(m)$如图2.20所示。使用三角恒等式2.178，我们可以将2.185写成如下形式

$$A(m_{\text{ML}}) = \left(\frac{1}{N}\sum_{n=1}^N\cos\theta_n\right)\cos\theta_0^{\text{ML}} - \left(\frac{1}{N}\sum_{n=1}^N\sin\theta_n\right)\sin\theta_0$$

2.187右边部分很容易评估，函数$A(m)$可以被数字化翻转。

为了完整性，我们简单设计一些重构周期分布的可替代技术。最简单的方法就是使用观测值的直方图，将极坐标划分成固定的bins。这很简单，也很灵活，但是还是会遇到一些限制，正如看到的我们会在2.5节更详细地讨论直方图方法。另一个方法，如von Mises分布，从在欧式空间上的高斯分布，边缘化到单位圆，而不是条件（Mardia和Jupp，2000年）。然而，这也会导致更加复杂的分布形式，我们也不会进一步讨论这些。最后，任何在实数轴上的可用分布（比如高斯）都可以转换成周期变量$（0, 2\pi）$，相应地会将实数轴微软这单位圆包裹起来。如果是这个结果分布比von Mises分布更加复杂的，那就更难处理了。

von Mises分布的一个限制就是单峰问题。通过形式化混合von Mises分布，我们得到更加灵活的周期变量的建模，它可以处理多峰问题。对于一个机器学习应用的例子，使用von Mises分布，可以看Lawrence等人（2002年），对于回归问题的条件密度的建模扩展，可以看Bishop和Nabney（1996年）。

#### 2.3.9 混合高斯

当高斯分布有着很多中啊哟的分析性质的时候，在对真实数据集建模时，它会遇到很多的限制。考虑如图2.21的例子。这被称为"Old Faithful"的大数据集，它记录了美国黄石国家公园Old Faithful喷泉的272次喷发的测量。每次测量都记录了喷发的持续时间（水平轴）和下次喷发的时间（垂直轴）。我们看到数据集包含两个主要的集合，这是简单的高斯分布是无法捕捉到这种结构的，然而，两个高斯的线性叠加，能够过更好地描述数据集。



<center>图2.21 图中画的蓝色曲线的常量概率密度等高线是"Old Faithful"数据。左边是最大似然拟合数据的单高斯分布。需要注意的是这个分布并不能捕捉数据中的两个内容，事实上，中间部分数据过于稀疏，而大部分高概率数据却不在这个位置。而右边分布，是两个高斯的线性组合，它们是第9章讨论的最大似然技术的线性组合，它给出了对数据的最好表达。</center>

这种叠加，是通过更多的简单分布，如高斯的线性组合，可以表示为被称为混合高斯（McLachlan和Basford，1988年；McLachlan和Peel，2000年）的概率模型。如图2.22中，我们看到高斯的线性组合，可以产生非常复杂的密度。通过对高斯数量的充分使用，通过调整他们的均值和协方差，同时在线性组合中的系数，几乎任何连续密度都可以近似得到任意精度。



<center>图2.22 一个维度的混合高斯分布的例子，展示了三个蓝颜色的高斯（每个都是系数的缩放），红色是它们的汇总。</center>

我们因此考虑一个$K$高斯密度的叠加的形式

$$p(\mathbf{x}) = \sum_{k=1}^K\pi_k\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}_k, \mathbf{\Sigma}_k)$$

这被称作混合高斯。每个高斯密度$\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}_k, \mathbf{\Sigma}_k)$都被称为混合的一部分，都有它各自的均值$\mathbf{\mu}_k$，协方差$\mathbf{\Sigma}_k$。如图2.23，等高线和混合高斯的表面有着3各组成部分。



<center>图2.23 图中画的是3个在二维空间的混合高斯。(a) 每个混合组合的常量密度等高线，有三个成员分别以红色，蓝色和绿色来表示，混合系数的值表示成如下每个成员。(b) 混合分布的边缘概率密度$p(\mathbf{x})$的等高线。(c) 画的是分布$p(\mathbf{x})$的表面。</center>

在这节，我们将考虑高斯组合来说明混合高斯框架。更一般地，混合模型是其它分布的线性组合。比如，在9.3.3解找那个，我们将考虑伯努利分布的混合，它是离散变量的混合模型的例子。

2.188中的参数$\pi_k$称为混合系数。对于$\mathbf{x}$，如果我们整合2.188的两边，需要注意的是对于$p(\mathbf{x})$和独立的高斯成员都是标准化过的，我们得到

$$\sum_{k=1}^K\pi_k = 1$$

同样有$p(\mathbf{x}) \ge 0$，$\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}_k, \mathbf{\Sigma}_k) \ge 0$，对于所有$k$，有$\pi_k \ge 0$。正例2.189的条件，我们有

$$0 \le \pi_k \le 1$$

我们因此看到混合高斯系数满足概率要求。

从加和和乘积规则中，边缘密度如下给出

$$p(\mathbf{x})=\sum_{k=1}^Kp(k)p(\mathbf{x} \mid k)$$

这与2.188等价，我们可以将$\pi_k = p(k)$看做是第$k^{\text{th}}$个成员的先验分布，密度$\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}_k, \mathbf{\Sigma}_k) = p(\mathbf{x} \mid k)$作为$\mathbf{x}$在条件$k$上的概率，这也成称为责任。从贝叶斯定理，可以给出如下

$$\begin{align*}
\gamma_k(\mathbf{x}) 
&\equiv p(k \mid \mathbf{x}) \\
&=\frac{p(k)p(\mathbf{x} \mid k)}{\sum_lp(l)p(\mathbf{x} \mid l)} \\
&=\frac{\pi_k\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}_k, \mathbf{\Sigma}_k)}{\sum_l\pi_l\mathcal{N}(\mathbf{x} \mid \mathbf{\mu}_l, \mathbf{\Sigma}_l)}
\end{align*}$$

我们将在第9章详细讨论混合分布的概率解释。

混合高斯的参数由参数$\mathbf{\pi}$，$\mathbf{\mu}$和$\mathbf{\Sigma}$表示，其中我们已经使用符号$\mathbf{\pi}\equiv\{\pi_1,\cdots,\pi_K\}$，$\mathbf{\mu}\equiv\{\mathbf{\mu}_1,\cdots,\mathbf{\mu}_K\}$，和$\mathbf{\Sigma}\equiv\{\mathbf{\Sigma}_1,\cdots,\mathbf{\Sigma}_K\}$。这些参数一种求解方法，就是使用最大似然。从2.188中得到对数似然函数如下

$$\ln p(\mathbf{X} \mid \mathbf{\pi}, \mathbf{\mu}, \mathbf{\Sigma}) = \sum_{n=1}^N\ln\left\{\sum_{k=1}^K\pi_k\mathcal{N}(\mathbf{x}_n \mid \mathbf{\mu}_K, \mathbf{\Sigma}_K)\right\}$$

其中$\mathbf{X} = \{\mathbf{x}_1,\cdots,\mathbf{x}_N\}$。我们很快就可以看到这种情形要比但个高斯要复杂很多，由于其内部存在对数，而且是在$k$个的综合。因此，参数的最大似然解也不会有闭合分析解。一种最大似然的方法，就是使用迭代数值优化技术（Fletcher，1987年；Nocedal和Wright，1999；Bishop和Nabney，2008年）。另外，我们会使用一个强大的框架叫做期望最大化，将在第9章讨论。
